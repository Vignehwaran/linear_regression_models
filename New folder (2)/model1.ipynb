```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("Salary_dataset.csv")

df.head()
df.info()
```

```python
%matplotlib  inline
plt.xlabel("YearofExperice")
plt.ylabel("Salary")
plt.scatter(df.YearsExperience,df.Salary,color="red")
```

```python
#create object the linear regression model 
reg=LinearRegression()
reg.fit(df[['YearsExperience']],df.Salary)
reg.predict([[10]])
```

Behind the scene

linearAlgebra 
y = mx + b, 
m=>slope or Gradient between the x and y axis
b=>y intetcept  


```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(df['YearsExperience'], df['Salary'], color='red')  # Scatter plot of the data
plt.plot(df['YearsExperience'], reg.predict(df[['YearsExperience']]), color='blue')  # Regression line
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary')
```
